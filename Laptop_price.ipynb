```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```python
df_1 = pd.read_csv('Laptop.csv', encoding='ISO-8859-1')
```

```python
df_1.head()
```

```python
df = df_1.drop(columns='laptop_ID',axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
label_encode = LabelEncoder()
```

```python
label_company = label_encode.fit_transform(df.Company)
label_product = label_encode.fit_transform(df.Product)
label_typename = label_encode.fit_transform(df.TypeName)
label_screenresolution = label_encode.fit_transform(df.ScreenResolution)
label_cpu = label_encode.fit_transform(df.Cpu)
label_ram = label_encode.fit_transform(df.Ram)
label_memory = label_encode.fit_transform(df.Memory)
label_gpu = label_encode.fit_transform(df.Gpu)
label_opsys = label_encode.fit_transform(df.OpSys)
label_weight = label_encode.fit_transform(df.Weight)
```

```python
df['Company'] = label_company
df['Product'] = label_product
df['TypeName'] = label_typename
df['ScreenResolution'] = label_screenresolution
df['Cpu'] = label_cpu
df['Ram'] = label_ram
df['Memory'] = label_memory
df['Gpu'] = label_gpu
df['OpSys'] = label_opsys
df['Weight'] = label_weight
```

```python
df.head()
```

```python
X = df.drop(columns='Price_euros',axis=1)
Y = df['Price_euros']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
X = standardized_data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train.shape , X_test.shape
```

```python
model = RandomForestRegressor(n_estimators=100)
```

```python
model.fit(X_train,Y_train)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
test_data_prediction
```

```python
error_score = metrics.r2_score(test_data_prediction,Y_test)
```
