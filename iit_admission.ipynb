```python
df_new = pd.read_csv('data-2.csv')
```

```python
df = df_new.drop(columns='Unnamed: 13',axis=1)
```

```python
label_encode = LabelEncoder()
```

```python
label_institute_type = label_encode.fit_transform(df.institute_type)
label_quota = label_encode.fit_transform(df.quota)
label_pool = label_encode.fit_transform(df.pool)
label_institute = label_encode.fit_transform(df.institute_short)
label_program = label_encode.fit_transform(df.program_name)
label_degree = label_encode.fit_transform(df.degree_short)
label_category = label_encode.fit_transform(df.category)
label_duration = label_encode.fit_transform(df.program_duration)
```

```python
df['institute_type'] = label_institute_type
df['institute_short'] = label_institute
df['quota'] = label_quota
df['category'] = label_category
df['degree_short'] = label_degree
df['program_name'] = label_program
df['pool'] = label_pool
df['program_duration'] = label_duration
```

```python
df.head()
```

```python
X = df.drop(columns=['institute_short','program_name','degree_short'])
Y = df[['institute_short','program_name','degree_short']]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(X_train,Y_train)
```

```python
test_data_prediction = model.predict(X_test)
```
