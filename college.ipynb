```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('kaggle_pivot_min_descending.csv')
```

```python
data.head()
```

```python
df = data.drop(columns=['mean','max-min','max-mean','sum'],axis=1)
```

```python
df
```

```python
label_encode = LabelEncoder()
```

```python
label_college = label_encode.fit_transform(df.college_name)
label_category = label_encode.fit_transform(df.seat_type)
label_branch = label_encode.fit_transform(df.branch)
label_type = label_encode.fit_transform(df.score_type)
```

```python
df['college_name'] = label_college
df['seat_type'] = label_category
df['branch'] = label_branch
df['score_type'] = label_type
```

```python
scaler = StandardScaler()
```

```python
X = df.drop(columns='college_name',axis=1)
Y = df['college_name']
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
standardized_data
```

```python
X = standardized_data
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train.shape , X_test.shape
```

```python
classifier = RandomForestClassifier(n_estimators=100)
```

```python
classifier.fit(X_train,Y_train)
```

```python
predicted_data = classifier.predict(X_test)
```

```python
predicted_data
```

```python
score = accuracy_score(Y_test,predicted_data)
score
```
