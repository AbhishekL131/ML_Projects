```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df = pd.read_csv('Train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# handling the missing values
# mean

mean_value = df['Item_Weight'].mean()
```

```python
df['Item_Weight'].fillna(mean_value,inplace=True)
```

```python
mode_value = df['Outlet_Size'].mode()
```

```python
mode_value
```

```python
df['Outlet_Size'].fillna('medium',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.head(5)
```

```python
label_encode = LabelEncoder()
```

```python
label_item = label_encode.fit_transform(df.Item_Identifier)
label_Fat = label_encode.fit_transform(df.Item_Fat_Content)
label_type = label_encode.fit_transform(df.Item_Type)
label_output = label_encode.fit_transform(df.Outlet_Identifier)
label_size = label_encode.fit_transform(df.Outlet_Size)
label_location = label_encode.fit_transform(df.Outlet_Location_Type)
label_outlet_type = label_encode.fit_transform(df.Outlet_Type)
```

```python
df['Item_Identifier'] = label_item
df['Item_Fat_Content'] = label_Fat
df['Item_Type'] = label_Fat
df['Outlet_Identifier'] = label_output
df['Outlet_Size'] = label_size
df['Outlet_Location_Type'] = label_location
df['Outlet_Type'] = label_outlet_type
```

```python
df.head()
```

```python
scaler = StandardScaler()
```

```python
X = df.drop(columns='Item_Outlet_Sales',axis=1)
Y = df['Item_Outlet_Sales']
```

```python
model = XGBRegressor()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train.shape , X_test.shape
```

```python
model.fit(X_train,Y_train)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
test_data_prediction
```

```python
error_score = metrics.r2_score(Y_test,test_data_prediction)
error_score
```
